## Permutations 
Given an array of distinct integers, return all the possible permutations. You can return the answer in any order.

```python
def all_permutations(a):
    
    if not a:
        return [[]]
    else:
        permutations = []
        l = len(a)
        if l == 1:
            permutations.append(a)
        elif l == 2:
            permutations.append([ a[0], a[1] ])
            permutations.append([ a[1], a[0] ])
        else:
            for i in range(l):
                p = all_permutations(a[0:i]+a[i+1:l])
                for j in p:
                    permutations.append([a[i]] + j )
    return permutations
```

```python
all_permutations([1,2,3])
```

```python
all_permutations([1,2,3,4,5])
```
