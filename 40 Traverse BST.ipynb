## Traverse BST
Write a 4 instance methods for a Binary Search Tree class to traverse the BST.<br>
1. Method 1 : traverse the tree breadth first and return an array that contains all the values of the BST.<br>
2. Method 2: traverse the tree depth first – In-order and return an array that contains all the values of the BST.<br>
3. Method 3 : traverse the tree depth first – Pre-order and return an array that contains all the values of the BST.<br>
4. Method 4 : traverse the tree depth first – Post-order and return an array that contains all the values of the BST.<br>

```python
#we used list

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


class BinarySearchTree:
    def __init__(self):
        self.root = None
    
    def bread_first(self):
        if not self.root:
            return
        else:
            array =[]
            queue = [self.root]
            while queue:
                node = queue.pop(0)
                array.append(node.value)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        return(array)
    
    def dfs_in_order(self):
        # inorder DFS, left node, cuurent node, right node
        if not self.root:
            return
        else:
            array =[]
            current = self.root

            def transverse(node):
                if node.left:
                    transverse(node.left)
                array.append(node.value)
                if node.right:
                    transverse(node.right)
            transverse(current)
            return array
        
    def dfs_pre_order(self):
        # inorder DFS, cuurent node, left node,  right node
        if not self.root:
            return
        else:
            array =[]
            current = self.root

            def transverse(node):
                array.append(node.value)
                if node.left:
                    transverse(node.left)
                if node.right:
                    transverse(node.right)
            transverse(current)
            return array
        
    def dfs_post_order(self):
        # inorder DFS, left node,  right node, cuurent node
        if not self.root:
            return
        else:
            array =[]
            current = self.root

            def transverse(node):
                if node.left:
                    transverse(node.left)
                if node.right:
                    transverse(node.right)
                array.append(node.value)
            transverse(current)
            return array
```
