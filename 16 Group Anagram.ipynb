## Group Anagrams 
Given an array of strings consisting of lower case English letters, group the anagrams together. You can return the answer in any order. An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, using all the original letters exactly once.

```python
def group_anagrams(a):
    dict = {}
    for word in a:
        w = ''.join(sorted(word))
        if w in dict:
            dict[w].append(word)
        else:
            dict[w] = [word]
    return (list(dict.values()))
```

```python
# how to make anagrams i.e permutations
def anagrams(a):
    
    if not a or len(a) == 1:
        return [a]
    else:
        permutations = []
        l = len(a)
        if l == 2:
            permutations.append( a[0] + a[1] )
            permutations.append( a[1] + a[0] )
        else:
            for i in range(l):
                p = anagrams(a[0:i]+a[i+1:l])
                for j in p:
                    permutations.append( a[i] + j )
    return permutations

anagrams("crtyujhg")
```
